```python
import pandas as pd
import numpy as np
import random
```

```python
#Abrir dataset de daots:
FILE_PATH = "../Data/phones.csv"
df = pd.read_csv(FILE_PATH, encoding='utf-8', error_bad_lines=False)
#df

#Veamos todas las columnas del dataset.
#df.columns

#Extraer columnas de interes:
df = df[['title','brand', 'rating', 'totalReviews','price']]
#df.head()

#Limpiamos el dataset:
df = df[df['price'] != 0.0]
df = df[df['price'] != 1.0]
#Quita marcas null o nan
df = df.dropna()
```

```python
df
```

```python
def chart_building(brand,max_price,df,percentile=0.8):

    #Define a new movies variable to store the preferred movies. Copy the contents of gen_df to movies
    #movies = gen_df.copy()

    #Filter based on the condition

    df = df[(df['brand'] == brand) & (df['price'] <= max_price)]

    #Compute the values of C and m for the filtered movies

    C = df['rating'].mean()

    m = df['totalReviews'].quantile(percentile)

    #Only consider phones that have higher than m votes. Save this in a new dataframe q_phone
    q_phone = df.copy().loc[df['totalReviews'] >= m]

    #Calculate score using the IMDB formula

    q_phone['score'] = q_phone['totalReviews']/(q_phone['totalReviews']+m) * q_phone['rating'] + (m/(m+q_phone['totalReviews']))*C

    #Sort phones in descending order of their scores

    q_phone = q_phone.sort_values('score', ascending=False)
    #print(q_phone['title'])
    #print(q_phone['score'])

    return q_phone
```

```python
INPUTS_PATH = './Inputs_knowledge/'
OUTPUTS_PATH = './Recommendations_knowledge/'
marca = list(df['brand'].unique())


resultado = []
for instance in range(100):
    input_file = open(INPUTS_PATH+'input_{}.txt'.format(instance),'w')
    output_file = open(OUTPUTS_PATH+'recommendation_{}.txt'.format(instance),'w')
    #Generate the chart for top animation movies and display top 5.
    #Numero aleatorio para marca:
    num1 = random.randint(0, (len(marca)-1))
    marca_r = marca[num1]
    input_file.write('Marca: {}\n'.format(marca_r)) 
    #print('marca es:'+marca_r)
    
    #Numero aleatorio para precio:
    precio_r = round(random.uniform (0, 1000),3)
    input_file.write('Precio: {} €\n'.format(precio_r))
    #print('precio es:',precio_r)
    
    KBS=chart_building(marca_r,precio_r,df)
    #KBS=chart_building(brand,max_price,df)

    #Obtenemos solo los nombres de los telefonos
    KBS=KBS.title

    if(KBS.size == 0):
        res = 'No existe telefono menor o igual a ese precio y de esa marca'
        #print("No existe telefono menor o igual a ese precio y de esa marca")
        output_file.write('Recomendacion: {}\n'.format(res))
    else:
        #Convertimos la lista de lo mejores 5 elementos en una lista 
        KBS = list(KBS)
        #Extraemos el primer elemento:
        KBS = KBS[0]
        output_file.write('Recomendacion: {}\n'.format(KBS))
        resultado.append(KBS)

```
