```python

class Application(ttk.Frame):
    def __init__(self, main_window):
        super().__init__(main_window)
        main_window.title("Sistema de recomendación de celulares")
        main_window.configure(width=200, height=400)
        
        # Valores por defecto en los campos de texto #
        self.PRICE_DEFAULT_VALUE = "Precio en €"

        # Expresión regular para verificar valores numéricos #
        self.DECIMAL_RE = re.compile(r"^\d*[.,]?\d*$")

        # Configuración de elementos del formulario #
        # Marca #
        self.brand_label = ttk.Label(main_window, text="Marca")
        self.brand_label.place(x=40, y=50)
        self.brand_combo = ttk.Combobox(main_window)
        self.brand_combo.place(x=100, y=50)
    
        marca = list(df['brand'].unique())
        marca.append('No tengo idea') 
        self.brand_combo["values"] = marca
        

        # Precio #
        self.price_label = ttk.Label(main_window, text="Precio")
        self.price_label.place(x=40, y=110)
        self.price_entry = ttk.Entry(main_window)
        self.price_entry.insert(tk.END, self.PRICE_DEFAULT_VALUE)
        self.price_entry.place(x=100, y=110)

        # Boton para buscar recomendacion #
        self.search_button = ttk.Button(main_window, text="Buscar", command=self.onClick_searchButton)
        self.search_button.place(x=230, y=170)
        
        #RESULTADO
        self.result_label = ttk.Label(main_window, text="Resultado")
        self.result_label.place(x=40, y=250)
        self.result = Text(main_window)
        self.result.place(x=150, y=250, height=130, width=250)

        main_window.configure(width=500, height=400)
        self.place(width=500, height=200)
        
    def chart_building(self,brand,max_price,df,percentile=0.8):

        #Define a new movies variable to store the preferred movies. Copy the contents of gen_df to movies
        #movies = gen_df.copy()

        #Filter based on the condition

        df = df[(df['brand'] == brand) & (df['price'] <= max_price)]

        #Compute the values of C and m for the filtered movies

        C = df['rating'].mean()

        m = df['totalReviews'].quantile(percentile)

        #Only consider phones that have higher than m votes. Save this in a new dataframe q_phone
        q_phone = df.copy().loc[df['totalReviews'] >= m]

        #Calculate score using the IMDB formula

        q_phone['score'] = q_phone['totalReviews']/(q_phone['totalReviews']+m) * q_phone['rating'] + (m/(m+q_phone['totalReviews']))*C

        #Sort phones in descending order of their scores

        q_phone = q_phone.sort_values('score', ascending=False)
        #print(q_phone['title'])
        #print(q_phone['score'])

        return q_phone

    def onClick_searchButton(self):
        #Marca
        marca = self.brand_combo.get()
        #print(marca)
        
        # Precio #
        price = self.price_entry.get()
        price = float(price)
        #print(price)
        
        if self.onCheck_Entries() == None:
            messagebox.showerror("Error", "valores invalidos")
        else:
            print("Entradas validas")
            
            if(marca == 'No tengo idea'):
                expert_window = tk.Toplevel(main_window)
                main_window.iconify()
                #main_window.destroy()
                #print(price)
                Expert2(expert_window, price)
            else:
                #CODIGO PARA ENCONTRAR TELÉFONO.
                
                #Generate the chart for top animation movies and display top 5.
                KBS=self.chart_building(marca,price,df).head()
                #KBS=chart_building(brand,max_price,df)

                #Obtenemos solo los nombres de los telefonos
                KBS=KBS.title
                
                if(KBS.size == 0):
                    messagebox.showerror("Error", "No tenemos ningun teléfono de esa marca con precio menor o igual al que escribiste")
                else:
                    #Convertimos la lista de lo mejores 5 elementos en una lista 
                    KBS = list(KBS)
                    #Extraemos el primer elemento:
                    KBS = KBS[0]

                    #KBS=KBS.reset_index(drop=True)
                    self.result.insert(INSERT, KBS)  

                
    def onCheck_Entries(self):
        # Precio #
        price = self.price_entry.get()

        # Se revisa que los campos de texto no tengan los valores por defecto o estén vacios #
        if price == self.PRICE_DEFAULT_VALUE or price == '':
            return None 
        if not self.DECIMAL_RE.match(price):
            return None

        return price

    


main_window = tk.Tk()
app = Application(main_window)

app.mainloop()
```
