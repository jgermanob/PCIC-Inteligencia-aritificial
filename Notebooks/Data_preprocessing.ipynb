# Preprocesamiento de datos

El objetivo de este notebook es realizar un preprocesamiento sobre el dataset, con el objetivo de poder utilizarlo en tareas de aprendizaje de m√°quina

### Bibliotecas

```python
import pandas as pd
import numpy as np
import re
```

### Carga del dataset

```python
FILE_PATH = "../Data/phone_dataset.csv"
```

```python
df = pd.read_csv(FILE_PATH, encoding='utf-8', error_bad_lines=False)
```

```python
df.head(10)
```

### Eliminaci√≥n de columnas no necesarias

```python
necessary_columns = ['brand', 'model', 'CPU', 'GPU', 
                     'internal_memory', 'RAM', 'primary_camera', 
                     'secondary_camera', 'battery', 'approx_price_EUR',
                     'colors', 'OS']
unused_columns = [column for column in df.columns if column not in necessary_columns]
```

```python
df = df.drop(unused_columns, axis=1)
```

#### Eliminaci√≥n de tel√©fonos que tienen NULL en alguna de sus caracter√≠sticas

```python
df = df.dropna(how='any',axis=0) 
```

```python
df
```

### Preprocesamiento de columna battery

A partir de la columna _battery_ es posible obtener 3 columnas, para identificar la siguiente informaci√≥n:
* Si la bateria es desmontable o no
* El tipo de bateria
* Los miliamperios de la bateria

```python
def removable(battery):
    if battery[0] == 'Removable':
        return True
    elif battery[0] == 'Non-removable':
        return False
    else:
        return np.nan
```

```python
a = df['battery'].astype(str).apply(lambda x: 
                                        (re.findall('(Non-removable|Removable|)\s*(Li-Ion|Li-Po|Li-ion|Li-po|NiMH|)\s*(\d*)',x))) 

df['battery_removable'] = a.apply(lambda x: removable(x[0])) # the first regex indicates 'removable' / 'non-removable'
df['battery_type'] = a.apply(lambda x: (x[0])[1]) # the second regex indicates 'Li-Ion' / 'Li-Po' / 'Li-ion' / 'Li-po' / 'NiMH'
df['battery_mah'] = a.apply(lambda x: ((x[0])[2]))# the third regex gives the numerical data for mah
```

```python
df
```

#### Eliminaci√≥n de la columna _battery_

```python
df = df.drop(columns=['battery'])
```

### Preprocesamiento de columna CPU

De esta columna, es posible obtener el n√∫mero de nucleos del procesador y la velocidad del mismo

```python
cpus = []
for cpu in df['CPU']:
    if cpu not in cpus:
        cpus.append(cpu)

for cpu in cpus:
    print(cpu)
```

#### Obtenci√≥n del n√∫mero de cores

```python
def get_cores(cpu):
    if cpu == 'Quad-core':
        return 4
    elif cpu == 'Octa-core':
        return 8
    elif cpu == 'Hexa-core':
        return 6
    elif cpu == 'Dual-core':
        return 2
    elif cpu == 'Triple-core':
        return 3
    elif cpu == 'Deca-core':
        return 10
    else:
        return 1
```

```python
a = df['CPU'].astype(str).apply(lambda x: 
                                        (re.findall('(Quad-core|Octa-core|Hexa-core|Dual-core|Triple-core|Deca-core|)\s*',x))) 

df['CPU_cores'] = a.apply(lambda x: get_cores(x[0]))
```

```python
df
```

#### Obtenci√≥n de velocidad del procesador

Con base en los valores de la columna _CPU_ es posible observar que la velocidad est√° dada en GHz o MHz. Con el objetivo de tener uniformidad en los datos, se utilizar√°n GHz como unidad.

```python
a = df['CPU'].astype(str).apply(lambda x: 
                                        (re.findall('(\d*.\d*)\s(GHz|MHz)',x)))
```

```python
a[a.apply(lambda x: len(x)==0)]
```

```python
def adjust_mhz(speed):
    return np.float64(speed)*0.001
```

```python
def get_cpu_speed(cpu):
    if len(cpu) > 0:
        cpu = cpu[0]
        if cpu[1] == 'GHz':
            return np.float64(cpu[0])
        elif cpu[1] == 'MHz':
            return adjust_mhz(cpu[0])
        else:
            return 0.0
    else:
        return 0.0
        
```

```python
df['CPU_speed'] = a.apply(lambda x: get_cpu_speed(x))
```
